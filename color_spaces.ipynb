```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = "Images/color_wheel.jpg"
img = cv2.imread(IMAGE_PATH,cv2.IMREAD_UNCHANGED)
plt.imshow(img)
plt.show()
```

```python
gray_frame = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_frame)
plt.show()
```

```python
luv_frame = cv2.cvtColor(img, cv2.COLOR_BGR2LUV)
plt.imshow(luv_frame)
plt.show()
```

```python
hsv_frame = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_frame)
plt.show()
```

```python
hsv_frame = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_frame)
plt.show()
```

```python
ycrcb_frame = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
plt.imshow(ycrcb_frame)
plt.show()
```
