```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
IMAGE_PATH = "Images/h_pipes_Big.jpg"
img = cv2.imread(IMAGE_PATH,cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(img)
plt.show()
```

Performing gaussian blur on the image with two seperate kernals.
Subtracting the blurred images to find the DoG.
(Difference of Gaussians)  

```python
gaussian1 = cv2.GaussianBlur(img, (9,9), 0)
gaussian2 = cv2.GaussianBlur(img, (7,7), 0)
```

```python
plt.imshow(gaussian1 - gaussian2)
plt.show()
```
